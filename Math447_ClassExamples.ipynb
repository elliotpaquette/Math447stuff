**Math 447 Example (Random walk dice rolls)**

A student asked how to generate plots like seen in the first lecture.  The plot I used in the lecture was a cartoon.  Here is a plot of the actual simulation.

-Elliot

```
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt

time=30
samples=100
cmap = plt.cm.get_cmap('hsv',samples)
for j in range(samples):
  v=npr.randint(1,7,size=time)
  for i in range(1,time):
    v[i]+=v[i-1]
  plt.plot(v,c=cmap(j))
```

Math 447 Example (Random walk on a graph)
In this example we actually simulate the states of a random walk on a 4-state Markov chain.
The transition matrix in this example is from the January 20, 2021 lecture notes.

```
import numpy as np
import numpy.random as npr
import numpy.linalg as npa
import matplotlib.pyplot as plt

alpha = [1,0,0,0]
P = np.matrix([[0,5/8,1/8,2/8],
               [5/7,0,0,2/7],
               [1,0,0,0],
               [1/2,1/2,0,0]
               ])

steps=100
states=np.zeros(steps,dtype=int)
states[0]=npr.choice(4,p=alpha)
for k in range(1,steps):
  states[k]=npr.choice(4,p=P.A[states[k-1]])
states+1

```

Math 447.  The following computes the 20-th power of a 4x4 transition matrix, specifically the example shown in class on January 20.

```
import numpy as np
import numpy.random as npr
import numpy.linalg as npa
import matplotlib.pyplot as plt

P = np.matrix([[0,5/8,1/8,2/8],
               [5/7,0,0,2/7],
               [1,0,0,0],
               [1/2,1/2,0,0]
               ])
npa.matrix_power(P,20)
```
